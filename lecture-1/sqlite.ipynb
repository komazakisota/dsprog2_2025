# データの永続化
## SQLite
* 軽量データベースを実装するC言語のライブラリ
* サーバー用のプロセスを必要としない、取り組み型のデータベースエンジン
* 単一のファイルによって管理される（インメモリも可能）

```python
import sqlite3
```

```python
# DBファイルの保存先パス(相対パスで指定)
path = ''

# Google Colabの場合は以下を指定
# path = '/content/'

# DBファイルのパス
db_name = 'test.db'

# DB接続の確立
conn = sqlite3.connect(path + db_name)

# DBへの接続を閉じる
conn.close()
```
